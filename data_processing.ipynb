```python
import csv
import pandas as pd
import math
import numpy as np

path = r'D:\Hoc\Dữ liệu\API binance\API.xlsx'        
my_data = pd.read_excel(path, header=0, sheet_name=None)
my_data
```

```python
nameOfAssets = list(my_data.keys())
numberOfAssets = len(nameOfAssets)
nameOfAssets.remove('FTT_USD')
nameOfAssets.remove('SHIB_USD')
# nameOfAssets.remove('LUNC_USD')
priceMatrix = []

timeStart = pd.to_datetime('2020-10-14').date()
timeEnd = pd.to_datetime('2022-05-6').date()

for nameAsset in nameOfAssets:
    dataAssets = my_data[nameAsset]
    dataAssets['Date'] = pd.to_datetime(dataAssets['Giờ mở giao dịch'], format='%Y-%m-%d').dt.date
    dataAssets =  dataAssets.loc[(dataAssets['Date'] >= timeStart)
                         & (dataAssets['Date'] <= timeEnd)]
    priceMatrix.append(dataAssets.iloc[:, 1].to_numpy())
print(nameOfAssets)    
print(len(nameOfAssets))  
```

```python
def percentPrice(priceArr, rangeTime):
    percentPriceList = []   
    for idx in range(len(priceArr) - 1, rangeTime, - rangeTime):  
        if priceArr[idx] == 0:
            break
        else:
            percentPrice = (priceArr[idx] - priceArr[idx - rangeTime])/priceArr[idx - rangeTime]
            percentPriceList.append(percentPrice)
    return percentPriceList

parameter = {'Day': 1, 'Week': 7, 'Month': 30}
def savePercentMatrix(priceMatrix, parameter):
    for key in parameter:
        percentPriceMatrix = []
        for price in priceMatrix:
            percentPriceArr = percentPrice(price, parameter[key])
            percentPriceMatrix.append(percentPriceArr)
        dataSave = pd.DataFrame(percentPriceMatrix).T
        dataSave.to_excel(excel_writer = f"D:\Hoc\Dữ liệu\API binance\percentReturns{key}DataBeforeLunc.xlsx")                
        
        
savePercentMatrix(priceMatrix, parameter)        
```
